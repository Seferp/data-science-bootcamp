````
import matplotlib
matplotlib.__version__
````

````
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````
fig = plt.figure()
fig.show()
````

### <a name='1'></a> Wykres liniowy

````
plt.plot([0,1], [0,3])
````

````
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres')
plt.legend()
````

````
plt.plot([0,1,2], [0,3,2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('wykres')
plt.legend()
````

````
x = np.arange(-3, 3, 0.1)
x_sin= np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
````

### <a name='2'> </a> Tworzenie wielu wykresów

````
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,2,5], 'r--')
````

````
fig = plt.figure(figsize=(12, 8))


plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,2,5], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,2,5], 'ro')
````

### <a name='3'> </a> Wykresy słupkowe


````
plt.bar(x=[0,1,2,3], height = [1.2,3.2,1.2,5.2])
````

````
plt.bar(x=['a','b','c','d'], height = [1.2,3.2,1.2,5.2])
````

````
plt.barh(y=['a','b','c','d'], width = [1.2,3.2,1.2,5.2])
````

### <a name='4'> </a> Wykresy punktowe

````
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
````

````
x3 = np.random.randn(300) * 50


plt.scatter(x=x1, y=x2, s=x3)
````

````
plt.scatter(x=x1, y=x2, s=x3, alpha = 0.5)
````
